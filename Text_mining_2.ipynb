```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
new_data = []
pat = r'^From:'
regexObj = re.compile(pat)
for i in lines:
    i = i.strip()
    matches = regexObj.findall(i)
    emails = re.findall("[0-9a-zA-z]+@[0-9a-zA-z]+\.[0-9a-zA-z]+",i)
    if matches != [] and len(emails)>0 and 40<len(i)<100:
        i = i.replace('From: ','')
        new_data.append(i)
```

```python
new_data
```

```python
names = []
emails = []
for i in new_data:
    email = re.findall("[0-9a-zA-z]+@[0-9a-zA-z]+\.[0-9a-zA-z]+",i)
    if re.findall(r'"(.*?)"',i) and len(email) > 0:
        names.append(re.findall(r'"(.*?)"',i))
        emails.append(email)
```

```python
emails = [i[0] for i in emails]
```

```python
names = [i[0] for i in names]
```

```python
dr_names = []
dr_emails = []
for i in range(len(names)):
    if re.search(r'^Dr.+',names[i],re.IGNORECASE) != None or re.search(r'^Dr+',names[i],re.IGNORECASE) != None:
        dr_names.append(names[i])
        dr_emails.append(emails[i])
```

```python
dr_emails
```

```python
dr_cn_names = []
dr_cn_emails = []
for i in range(len(dr_names)):
    if re.search(r'\.com$',dr_emails[i]) or re.search(r'\.net$',dr_emails[i]):
        dr_cn_emails.append(dr_emails[i])
        dr_cn_names.append(dr_names[i])
```

```python
dr_cn_emails
```

```python
domains = []
for i in dr_cn_emails:
    a = re.split('@',i)
    domains.append(a[1])
```

```python
df = pd.DataFrame(list(zip(dr_cn_names,dr_cn_emails,domains)),columns = ['Name','E-mail Address','E-mail Domain'])
```
